```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
import calendar
import warnings
warnings.simplefilter('ignore')
```

```python
covid = pd.read_csv('./us-counties.csv')
max(covid.date)
covid.head()
```

```python
covid['year']=covid['date'].str[0:4]
covid['month']=covid['date'].str[5:7]
covid['mo_yr']=covid['month']+"/"+covid['year']
covid['month'] = pd.to_numeric(covid['month'])
covid['year'] = pd.to_numeric(covid['year'])
covid['days_in_mo'] = covid['mo_yr'].apply(lambda t: pd.Period(t, freq='S').days_in_month)
```

```python
covid['new_col'] = covid['year'].astype(str)+'-'+covid['date'].str[5:7]+'-'+covid['days_in_mo'].astype(str)
covid['last_flag']= covid['date']==covid['new_col']
```

```python
covid_df = covid[covid['last_flag']].copy()
```

```python
covid_df['co_st'] = covid_df['county']+", "+covid_df['state']
```

```python
len(covid_df['co_st'].unique())
```

```python
covid_df.drop('co_st',axis=1,inplace=True)
```

```python
abbrev = pd.read_csv('abbrev.csv')
abbrev.head()
```

```python
covid_df = pd.merge(covid_df,abbrev,left_on='state',right_on='State')
```

```python
covid_df.drop(['State','Abbrev'],axis=1,inplace=True)
```

```python
covid_df['co_st'] = covid_df['county'] + ", " + covid_df['Code']
covid_df
```

```python
covid_df.to_csv('covid_df.csv')
```
