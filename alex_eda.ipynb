```python
covid = pd.read_csv('./us_counties_covid19_daily.csv')

# covid['year']=x['date']
# covid['month']=x

covid['year']=covid['date'].str[0:4]
covid['month']=covid['date'].str[5:7]
covid['mo_yr']=covid['month']+"/"+covid['year']

import calendar
calendar.monthrange(2020,1)


covid['month'] = pd.to_numeric(covid['month'])
covid['year'] = pd.to_numeric(covid['year'])
# calendar.monthrange(
calendar.monthrange(2019, 2)[1] # num_day
covid['day_in_mo'] = covid['mo_yr'].apply(lambda t: pd.Period(t, freq='S').days_in_month)
covid
```

```python
covid['new_col'] = covid['year'].astype(str)+'-'+covid['date'].str[5:7]+'-'+covid['day_in_mo'].astype(str)
covid['last_flag']= covid['date']==covid['new_col']
covid_df = covid[covid['last_flag']]
```
