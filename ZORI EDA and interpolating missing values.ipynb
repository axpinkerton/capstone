```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
zori = pd.read_csv('./zillow_data/ZORI_AllHomesPlusMultifamily_SSA_2021-06.csv', index_col=0)
zori_df = zori.copy()
```

```python
zori.shape
```

```python
zori_nulls = zori.isnull().sum(axis=0)
zori_nulls = zori_nulls[zori_nulls > 0].sort_values(ascending= False)
zori_nulls = pd.DataFrame(zori_nulls, columns = ['Nulls'])
zori_nulls['Yr_Mo'] = zori_nulls.index
zori_nulls.tail()
```

```python
zori_nulls = zori_nulls.merge(zori_nulls.Yr_Mo.apply(lambda x: pd.Series({"Year": int(x[0:4]), "Month": int(x[5:])})),
                 left_index=True, right_index=True)
zori_nulls['PercentNull'] = np.round(zori_nulls['Nulls']/zori.shape[0], 2)#.describe()
zori_nulls.drop('Yr_Mo', axis=1, inplace=True)
```

```python
zori_year_nulls = zori_nulls.groupby(['Year']).agg({'Nulls': np.sum, 'PercentNull': np.mean}).reset_index()
zori_year_nulls
```

```python
zips_ln = zori[zori['2014-01'].isnull()]
zips_en = zori[zori['2021-05'].isnull()]

print(zips_ln.shape)
print(zips_en.shape)

zips_ln.head()
```

```python
info_cols = ['SizeRank', 'MsaName']

print(zori.shape)
zori_df = zori.copy()
zori_df.drop(info_cols, axis=1, inplace=True)
zori_df.head()
print(zori_df.shape)

zori_melt = pd.melt(zori_df, id_vars=['RegionName'], var_name='Yr_Mo', value_name='RI')

zori_long = zori_melt.merge(zori_melt.Yr_Mo.apply(lambda x: pd.Series({"Year": int(x[0:4]), "Month": int(x[5:])})),
                 left_index=True, right_index=True)
zori_long.head()
```

```python
zori_long.sort_values(by=['RegionName', 'Year', 'Month'], inplace=True)
zori_long_copy = zori_long.copy()
zori_long
```

```python
zori_long[zori_long['RegionName']=='11937']
```

```python
for i in range(len(zori_long_copy.RegionName.unique())):
    mask = zori_long_copy.loc[:,'RegionName']==zori_long_copy.RegionName.unique()[i]
    zori_long_copy[mask]= zori_long_copy[mask].interpolate(method='linear')
```

```python
# Getting a value set for 2014-01 null RIs, so that data can be interpolated backwards

for i in range(len(zori_long_copy.RegionName.unique())):
    mask = zori_long_copy.loc[:,'RegionName']==zori_long_copy.RegionName.unique()[i]
    if(zori_long_copy[mask].RegionName.unique() in zips_ln['RegionName'].values):
#        print(zori_long[mask].head(1))        
        region = zori_long[mask].RegionName.unique()[0]
        series = zori_long[mask][zori_long[mask]['Year']==2014]['RI']
        index = series.index[0]
        diff = series.diff()
        diff = diff.replace(np.nan, 0)
        mean = sum(diff.values)/len(diff)
        num_nulls = series.isnull().sum()
        RI_chng = series.values[num_nulls] - (num_nulls * mean)
#        print(RI_chng)
        zori_long.loc[index, 'RI'] = RI_chng
        
```

```python
zori_long
```

```python
# Interpolate backwards
for i in range(len(zori_long.RegionName.unique())):
    mask = zori_long.loc[:,'RegionName']==zori_long.RegionName.unique()[i]
    zori_long[mask]= zori_long[mask].interpolate(method='linear', limit_direction='backward')
```

```python
# Quicksave
zori_lead_nulls_fixed = zori_long.copy()
```

```python
# Setting a value for 2021-05 null RIs, so that data can be interpolated forwards

for i in range(len(zori_long.RegionName.unique())):
    mask = zori_long.loc[:,'RegionName']==zori_long.RegionName.unique()[i]
    if(zori_long[mask].RegionName.unique() in zips_en['RegionName'].values):
        print(zori_long[mask].tail(3))        
        region = zori_long[mask].RegionName.unique()[0]
        series = zori_long[mask][zori_long[mask]['Year']==2021]['RI']
        s_size = series.size
#        print(s_size)
        index = series.index[s_size-1]
        diff = series.diff()
        diff = diff.replace(np.nan, 0)
        mean = sum(diff.values)/len(diff)
        num_nulls = series.isnull().sum()
        RI_chng = series.values[s_size-num_nulls-1] + (num_nulls * mean)
#        print(RI_chng)
        zori_long.loc[index, 'RI'] = RI_chng
        print(zori_long.loc[index, 'RI'])
        print(zori_long[mask].tail(3), "\n")
```
