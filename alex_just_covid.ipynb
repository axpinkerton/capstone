```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
import warnings
warnings.simplefilter('ignore')
import calendar
```

```python
def fix_zip(series):
      return series.astype(str).str.replace('[^\w\s]', '').str.zfill(5)
```

```python
covid = pd.read_csv('./us_counties_covid19_daily.csv')
covid['year']=covid['date'].str[0:4]
covid['month']=covid['date'].str[5:7]
covid['mo_yr']=covid['month']+"/"+covid['year']
covid['month'] = pd.to_numeric(covid['month'])
covid['year'] = pd.to_numeric(covid['year'])
covid['days_in_mo'] = covid['mo_yr'].apply(lambda t: pd.Period(t, freq='S').days_in_month)
covid
```

```python
covid['new_col'] = covid['year'].astype(str)+'-'+covid['date'].str[5:7]+'-'+covid['days_in_mo'].astype(str)
covid['last_flag']= covid['date']==covid['new_col']
covid_df = covid[covid['last_flag']]
```

```python
covid_df.sort_values(by=['state','county','date'],inplace=True)
```

```python
covid_df['new_cases']=covid_df['cases']-covid_df['cases'].shift(1)
covid_df['new_deaths']=covid_df['deaths']-covid_df['deaths'].shift(1)
abbrev = pd.read_csv('./abbrev.csv')
abbrev.head()
#  covid_df.to_csv('./covid_csv.csv')
covid_df = pd.merge(covid_df,abbrev,left_on='state',right_on='State')
covid_df['co_st']=covid_df['county']+', '+covid_df['Code']
```

```python
county_to_zip=pd.read_csv('zillow_geog.csv')
county_to_zip.columns = ['RegionID', 'SizeRank', 'ZIP', 'RegionType', 'StateName',
       'State', 'City', 'Metro', 'County']
county_to_zip['co_st'] = county_to_zip['County'].str[:-7]
county_to_zip['co_st'] = county_to_zip['co_st']+', '+county_to_zip['State']
county_to_zip['ZIP']=fix_zip(county_to_zip['ZIP'])
county_to_zip

county_to_zip[['ZIP','co_st']]
```

```python
census = nl[['GEO_ID','DP05_0001E']]
census.drop(0,inplace=True,axis=0)
```

```python
census['zip'] = census['GEO_ID'].str[-5:]
census
```

```python
census.columns=['geo_id','ttl_pop','zip']
zip_co = pd.merge(census,county_to_zip[['ZIP','co_st']],left_on='zip',right_on='ZIP')
zip_co.drop('ZIP',axis=1,inplace=True)
zip_co['ttl_pop']=zip_co['ttl_pop'].astype(int)
ttl_county_pop = zip_co.groupby('co_st')[['ttl_pop']].agg(sum)
ttl_county_pop.reset_index(inplace=True)
ttl_county_pop.columns=['county_st',"county_pop"]
county_zip_df = pd.merge(zip_co,ttl_county_pop,left_on='co_st',right_on='county_st')
county_zip_df['percent_to_ttl']=county_zip_df['ttl_pop']/county_zip_df['county_pop']

# county_zip_covid = pd.merge(covid_df,county_zip_df,left_on='co_st',right_on='co_st')
# county_zip_covid.to_csv('county_zip_covid.csv')
```

```python
county_zip_covid=pd.read_csv('./county_zip_covid.csv')
```

```python
county_zip_covid['yr_mo']=county_zip_covid['date'].str[0:7]
```

```python
zori = pd.read_csv('./Zip_ZORI_AllHomesPlusMultifamily_SSA.csv')
zori['RegionName']=fix_zip(zori['RegionName'])

zori_zips = list(zori['RegionName'].unique())
```

```python
filtered_covid = county_zip_covid[county_zip_covid['zip'].isin(zori_zips)]
filtered_covid
```

```python
filtered_covid['yr_mo']=filtered_covid['date'].str[0:7]
filtered_covid['zip_cases'] = (filtered_covid['cases']*filtered_covid['percent_to_ttl']).astype(int)
filtered_covid['zip_new_cases']=(filtered_covid['new_cases']*filtered_covid['percent_to_ttl']).astype(int)
filtered_covid['zip_deaths'] = (filtered_covid['deaths']*filtered_covid['percent_to_ttl']).astype(int)
filtered_covid['zip_new_deaths']=(filtered_covid['new_deaths']*filtered_covid['percent_to_ttl']).astype(int)

filtered_covid['zip_cases_to_ttl']=filtered_covid['zip_cases']/filtered_covid['ttl_pop']
filtered_covid['zip_new_cases_to_ttl']=filtered_covid['zip_new_cases']/filtered_covid['ttl_pop']
filtered_covid['zip_deaths_to_ttl']=filtered_covid['zip_deaths']/filtered_covid['ttl_pop']
filtered_covid['zip_new_deaths_to_ttl']=filtered_covid['zip_new_deaths']/filtered_covid['ttl_pop']
cleaned_covid = filtered_covid[['yr_mo','zip','zip_cases_to_ttl','zip_new_cases_to_ttl','zip_deaths_to_ttl','zip_new_deaths_to_ttl']]
cleaned_covid
cleaned_covid.to_csv('cleaned_covid.csv')
```

```python
Xa = cleaned_covid[['yr_mo','zip','zip_cases_to_ttl']].pivot_table(columns='zip',index='yr_mo')
Xb = cleaned_covid[['yr_mo','zip','zip_new_cases_to_ttl']].pivot_table(columns='zip',index='yr_mo')
Xc = cleaned_covid[['yr_mo','zip','zip_deaths_to_ttl']].pivot_table(columns='zip',index='yr_mo')
Xd = cleaned_covid[['yr_mo','zip','zip_new_deaths_to_ttl']].pivot_table(columns='zip',index='yr_mo')

Xa.columns = Xa.columns.droplevel(0)
Xa.reset_index(inplace=True)
Xb.columns=Xb.columns.droplevel(0)
Xb.reset_index(inplace=True)
Xc.columns=Xc.columns.droplevel(0)
Xc.reset_index(inplace=True)
Xd.columns=Xd.columns.droplevel(0)
Xd.reset_index(inplace=True)

Xa
```
